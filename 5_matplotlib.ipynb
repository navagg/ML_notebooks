```python
def f(t): 
    return np.exp(-t) * np.cos(2*np.pi*t) 
t1 = np.arange(0.0, 5.0, 0.1) 
t2 = np.arange(0.0, 5.0, 0.02) 
plt.figure(1) # Called implicitly
# for multiple figures 
plt.subplot(211) # 2 rows, 1 column, 1st plot 
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k') 
plt.subplot(212) # 2 rows, 1 column, 2nd plot 
plt.plot(t2, np.cos(2*np.pi*t2), 'r--') 
plt.show() 
#%%
```

##### Create a line graph of CO2 versus time. View the plot.

```python
times = range(7)
co2 = [250, 265, 272, 260, 300, 320, 389]
plt.plot(times, co2)
plt.plot(times, co2, 'b--')
plt.title("Concentration of CO2 versus time")
plt.ylabel("[CO2]")
plt.xlabel("Time (decade)")
plt.show()
```

##### Continuing with the above data plot, add some additional data

```python
temp = [14.1, 15.5, 16.3, 18.1, 17.3, 19.1, 20.2]
plt.plot(times, co2, 'b--', times, temp, 'r*-')
plt.show()
plt.savefig("co2_temp.pdf")
```

##### Multiple axes

```python
fig, ax1 = plt.subplots()
#times = range(7)
#>>> co2 = [250, 265, 272, 260, 300, 320, 389]
ax1.plot(times, co2, "b--")
ax1.set_ylabel("[CO2]")
ax2 = ax1.twinx()
temp = [14.1, 15.5, 16.3, 18.1, 17.3, 19.1, 20.2]
ax2.plot(times, temp, "r*-")
ax2.set_ylabel("Temp (degC)")
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
mu, sigma = 100, 15 
x = mu + sigma * np.random.randn(10000) 
# the histogram of the data 
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75) 
plt.xlabel('Smarts') 
plt.ylabel('Probability') 
plt.title('Histogram of IQ') 
plt.text(60, .025, '$\mu=100,\ \sigma=15$', color='r') #TeX equations 
plt.axis([40, 160, 0, 0.03]) 
plt.grid(True) 
plt.show() 
```

```python
import scipy
import matplotlib.pyplot as plt
import scipy.io.wavfile
sample_rate, X = scipy.io.wavfile.read('congestion.wav')
#print (sample_rate, X.shape )
plt.specgram(X, Fs=sample_rate, xextent=(0,30))
plt.colorbar()
plt.show()
```

```python
import matplotlib.pyplot as plt
import librosa as lb
import librosa.display
y, sr = lb.load('congestion.wav')
plt.figure(figsize=(12, 8))
D= lb.amplitude_to_db(lb.stft(y), ref=np.max)
plt.subplot(4, 2, 1)
lb.display.specshow(D, y_axis='linear')
plt.colorbar()
plt.title('Linear-frequency power spectrogram')
plt.show()
```
